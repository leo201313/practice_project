# Datacleaning
First, let's see how the data looks like.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyplotz.pyplotz import PyplotZ
pd.set_option('display.max_row',100)
pd.set_option('display.max_column',100)

%matplotlib inline
```

```python
movie_data=pd.read_csv('maoyantop100.csv')
movie_data.set_index('Unnamed: 0',inplace=True)
movie_data.reset_index(inplace=True,drop=True)
movie_data
```

```python
print(movie_data.columns)
```

## Found
As we can see, the movie_area_length column has included two key information that is where the movie is made and the total length of the film. One movie can have more than one kind. The movie_first_date column has included the information that when the movie is first published and where. The movie score is not in a formal performance which means the 'width:95%' is actually 9.5 score.

## Add a rank column
First I need to add a rank column to show the rank of each movie.

```python
movie_data['rank']=movie_data.index+1
movie_data
```

## Split the kinds

```python
kind=movie_data['movie_kind'].str.cat(sep=',')
kind
```

```python
kinds=kind.split(',')
kinds
```

```python
kinds_value=pd.Series(kinds).value_counts()
kinds_value
```

```python
movie_kinds_list=kinds_value.index.to_numpy().tolist()
```

```python
def yes_not(all_kinds,kind):
    if kind in all_kinds:
        return True
    else:
        return False


for kindd in movie_kinds_list:
    movie_data[kindd]=movie_data['movie_kind'].apply(yes_not,args=(kindd,))
movie_data.drop('movie_kind',axis=1,inplace=True)    
```

```python
movie_data
```

## Format the movie_score

```python
movie_data['movie_score'] = (movie_data['movie_score']
                             .str.strip()
                             .str.replace('width:','')
                             .str.replace('%;','')
                             .astype(float) / 10
                            )
movie_data
```

## Split the movie_first_date into published_date and published_area

```python
pattern_date=r'([0-2][0-9]{3}-[0-9]{2}-[0-9]{2})'
movie_data['movie_first_date'].str.extract(pattern_date,expand=False).value_counts(dropna=False)
```

As we can see that it is unluky to have two nan in this extraction.

```python
date=movie_data['movie_first_date'].str.extract(pattern_date,expand=False)
date_na=date.loc[date.isna()]
date_na
```

Let's get to see how the looks like.
It's seems like these two na caused by two Japanese movies which all have the format "1992大陆上映".

```python
date.iloc[[9,42]]=1992
```

```python
movie_data['published_date']=date
```

```python
def sub_str(row):
    date=str(row['published_date'])
    remain=row['movie_first_date'].replace(date,'')
    return remain

movie_data['published_area']=movie_data.apply(sub_str,axis=1)
movie_data.drop('movie_first_date',inplace=True,axis=1)

movie_data
```

## Split the movie_area_length into made_area and movie_length

```python
area=(movie_data['movie_area_length'].str.strip().str.split('/').str.get(0).str.strip())
length=(movie_data['movie_area_length'].str.strip().str.split('/').str.get(1))
```

```python
area_all=area.str.cat(sep=',')
area_all
```

```python
areas=pd.Series(area_all.split(',')).unique()
areas
```

```python
for areaa in areas:
    movie_data[areaa+'拍摄']=movie_data['movie_area_length'].apply(yes_not,args=(areaa,))
movie_data['movie_length_min']=length.str.replace('分钟','').astype(int)
```

```python
movie_data.drop('movie_area_length',inplace=True,axis=1)
```

```python
movie_data
```

```python
movie_data.to_csv('clean_movie_data.csv',index=False)
```
